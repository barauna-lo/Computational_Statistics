# 2.Regression - Inverse Problem

*   From x, y(x), what is the value of 'a' and 'b'?



It is known that given two points, that is, $A(x_0, y_0)$ and $B(x_1, y_1)$, it is possible to find the values of the angular ($a$) and linear ($b$) coefficients. This section shows two ways to do this.

## 2.1. Using the formulas of $a$ and $b$

To calculate the angular coefficient ($a$) of a line from two points $A (x_{0},y_{0})$ and B $(x_{1},y_{1})$ we must divide the variation between the axes as follows:

$a = \frac{\Delta_{y}}{\Delta_{x}} = \frac{y_{1}-y_{0}}{x_{1}-x_{0}}$

Once you find the value of $a$, just replace that value into the equation of the line and find the value of $b$.

$b = y_{0}-ax_{0}$

**Description of code:**

The function below finds the values of $a$ and $b$ given two points. In order to facilitate data manipulation, consider each point as an array called 'point' of two values, $point[0]=x$ and $point[1]=y$, respectively.



```python
#Description: Finds the values of 'a' and 'b', considering 'y = ax + b'.
#Input: Takes as parameter two points: point1=[x1,y1], point2=[x2,y2].
#Output: Returns the values of 'a' and 'b'.
def find_a_b(point1,point2):
  a = (point2[1]-point1[1]) / (point2[0]-point1[0])
  b = point1[1] - a*point1[0]
  return a, b
```

The code below prints the result of the 'find_a_b' function for two examples:

*   Example 1: $point_1(1,2)$ and $point_2(2,4)$
*   Example 2: $point_1(5,0)$ and $point_2(3,9)$



```python
#Example of using the 'find_a_b' function:

#2 = a + b
#4 = 2a + b
a1, b1 = find_a_b ([1,2], [2,4])
print("Result for the points (1,2), (2,4):", "\n\ta: ", a1, "\n\tb: ", b1)

#0 = 5a + b
#9 = 3a + b
a2, b2 = find_a_b ([5,0], [3,9])
print("Result for the points (5,0), (3,9):", "\n\ta: ", a2, "\n\tb: ", b2)
```

## 2.2. Solving a Linear System

Another way to find the values of $a$ and $b$ is by solving a linear system given the equation of the line.

**Description of code:**

The code below defines a function that solves a linear system of two variables given two points with the same structure as above.

```python
#Description: Solve a linear system, considering 'y = ax + b'
#Input: Takes as parameter two points: point1=[x1,y1], point2=[x2,y2]
#Output: Returns an array [a,b]
def solve_linear_system(point1,point2):
  S = np.array([[point1[0], 1.],
              [point2[0], 1.]])
  y = np.array([point1[1], point2[1]])
  return np.linalg.solve(S, y)
```

**Description of code:**

The code below prints the result of the 'solve_linear_system' function for the same two previous examples.


```python
#Example of using the 'solve_linear_system' function:

#2 = a + b
#4 = 2a + b
result1 = solve_linear_system ([1,2], [2,4])
print("Result for the points (1,2), (2,4):", "\n\ta: ", "{:.2f}".format(result1[0]), "\n\tb: ", "{:.2f}".format(result1[1]))

#0 = 5a + b
#9 = 3a + b
result2 = solve_linear_system ([5,0], [3,9])
print("Result for the points (5,0), (3,9):", "\n\ta: ", "{:.2f}".format(result2[0]), "\n\tb: ", "{:.2f}".format(result2[1]))
```

**Conclusion:**

*   It is possible to find the values of $a$ and $b$ from the equation of the line with only two points.

### 3.3.1. Exhaustive Search

*   $y(x)$, for a given $a$ and $b$
*   Error on a parameter space - exhaustive search for $a$ and $b$

The exhaustive search evaluates all possible parameters $(a, b)$ belonging to the parameter space.

For each model $(a, b)$, the values of $x$ according to the domain of $x$ are tested and the error is calculated. Then an average of the errors found for each model is made.

The winning model $(a, b)$ is the one with the lowest mean error.

**Description of code:**

For this work, three types of average errors were considered:

1.   Mean Error (ME): $\frac{\sum_{i=1}^{n} (f_{real(x_i)} - f_{estimated(x_i)})}{n}$
2.   Mean Absolute Error (MAE): $\frac{\sum_{i=1}^{n} |f_{real(x_i)} - f_{estimated(x_i)}|}{n}$
3.   Mean Squared Error (MSE): $\frac{\sum_{i=1}^{n} (f_{real(x_i)} - f_{estimated(x_i)})^2}{n}$

The code below defines several functions that support the calculation of these three types of errors, and also defines an array and a matrix of errors.

```python
#Linear function for a given 'a' and 'b'
def f_estimated(x, a, b): 
  return a*x + b

def calc_error_diff(x, a, b):
  return f_real(x) - f_estimated(x, a, b)

def calc_error_diff_abs(x, a, b):
  return np.abs(f_real(x) - f_estimated(x, a, b))

def calc_error_diff_sqrt(x, a, b):
  return (f_real(x) - f_estimated(x, a, b))**2    

def calc_average_error_diff(x_domain, a, b):
  error_array = []
  for x in x_domain:
    error_array.append(calc_error_diff(x, a, b))
  return np.mean(error_array)

def calc_average_error_diff_abs(x_domain, a, b):
  error_array = []
  for x in x_domain:
    error_array.append(calc_error_diff_abs(x, a, b))
  return np.mean(error_array) 

def calc_average_error_diff_sqrt(x_domain, a, b):
  error_array = []
  for x in x_domain:
    error_array.append(calc_error_diff_sqrt(x, a, b))
  return np.mean(error_array)       

#Calculate an array of errors
def calc_error_array (x_domain, array1, array2, calc_average_error):
  error_values = []
  for i in range(0, len(array1)):
    error_values.append(calc_average_error(x_domain, array1[i], array2[i]))
  return error_values  

#Calculate a matrix of errors
def calc_error_matrix (x_domain, arrayLine, arrayColumn, calc_average_error):
  #Create the matrix with the desired dimensions
  error_values = np.ndarray((len(arrayLine),len(arrayColumn)))
  #Populates the matrix with the respective errors, where error_values[i][j] is the average error of the model (arrayLine[i], arrayColumn[j])
  for i in range(0, len(arrayLine)):
    for j in range(0, len(arrayColumn)):
      error_values[i][j] = calc_average_error(x_domain, arrayLine[i], arrayColumn[j])
  return error_values    

#Find the best model in a matrix
def calc_best_model_matrix(error_matrix, arrayLine, arrayColumn):
  best_model = [arrayLine[0],arrayColumn[0]]
  error_min = error_matrix[0][0]
  for i in range(0, len(arrayLine)):
    for j in range(0, len(arrayColumn)):
      if error_matrix[i][j] < error_min:
        error_min = error_matrix[i][j]
        best_model = [arrayLine[i],arrayColumn[j]]
  return error_min, best_model  

#Find the best model in an array
def calc_best_model_array(error_array, array1, array2):
  pos_min = np.argmin(error_array)
  return error_array[pos_min], array1[pos_min], array2[pos_min]  

def calc_exhaustive_search(x_domain, array1, array2, calc_average_error):
  error_values = calc_error_array(x_domain, array1, array2, calc_average_error)
  error_min, a_best, b_best = calc_best_model_array(error_values, array1, array2)
  return error_values, error_min, a_best, b_best
```

**Description of code:**

The code below calculates and prints the best model for the exhaustive search considering the Mean Absolute Error (MAE).

The best model is the one with the lowest Mean Absolute Error (MAE).

In addition, the execution time of the exhaustive search is stored in the variable 'time_exhaustive_search'.

```python
start = time.time()
#Find the best model for the Exhaustive Search with the Mean Absolute Error (MAE)
error_values, error_min, a_best, b_best = calc_exhaustive_search(x_domain, a_values, b_values, calc_average_error_diff_abs)
end = time.time()
time_exhaustive_search = end - start

print("Array with the values of 'a':\n", a_values)
print("Array with the values of 'b':\n", b_values)
print("Array of Errors:\n", error_values)

print("\n\n***Exhaustive Search***\n")
print("Min. Error: ", error_min)
print("Best model: ", "\n\ta: ", a_best, "\n\tb: ", b_best)
print("Execution Time (seconds): ", time_exhaustive_search)
```

### 3.3.2. Challenge 1: Random Search


**Description of code:**

The code below considers **steps** as steps in a random search, and calculates the error at each step according to f_calc_average, which can be:

1.   Mean Error (ME)
2.   Mean Absolute Error (MAE)
3.   Mean Squared Error (MSE)

The model with the lowest average error is considered the winner.

```python
def cal_random_search(f_calc_average, steps, seed):

  generator = Generator(PCG64(seed))
  a_steps=[]
  b_steps=[]
  diff_array=[]

  for i in range(0, steps):
    a_step_i = a_domain[generator.integers(low=0, high=len(a_domain))]
    a_steps.append(a_step_i)
    b_step_i = b_domain[generator.integers(low=0, high=len(b_domain))]
    b_steps.append(b_step_i)
    #Calculate the Mean 
    diff_array.append(f_calc_average(x_domain, a_step_i, b_step_i))

  return diff_array, a_steps, b_steps
```

```python
#Define the number of iterations
steps = 500
#Define the seed
seed = 1

#Calculate the Mean Error (ME)
error_diff_array, a_steps_diff, b_steps_diff = cal_random_search(calc_average_error_diff, steps, seed)
#Calculate the Mean Absolute Error (MAE)
error_diff_abs_array, a_steps_diff_abs, b_steps_diff_abs = cal_random_search(calc_average_error_diff_abs, steps, seed)
#Calculate the Mean Squared Error (MSE)
error_diff_sqrt_array, a_steps_diff_sqrt, b_steps_diff_sqrt = cal_random_search(calc_average_error_diff_sqrt, steps, seed)
```

# 4.Noisy Data

*   Noise: additive, multiplicative, additive and multiplicative
*   How to control the noise intensity?
*   BTW, what is a noise?

The following types of noise were used in this work:

1.   Additive Noise: $f(x) = (ax + b) + r$
2.   Multiplicative Noise: $f(x) = (ax + b) * r$
3.   Additive and Multiplicative Noise: $f(x) = (ax + b) * r1 + r2$

The noise intensity was modeled as:

*   $\delta$ | $0 \leq \delta \leq 1$

For example, we can simulate the data with 30% noise ($\delta = 0.3$).

What is a noise?

Noisy data are data with a large amount of additional meaningless information called noise. This includes data corruption, and the term is often used as a synonym for corrupt data. It also includes any data that a user system cannot understand and interpret correctly. Many systems, for example, cannot use unstructured text. Noisy data can adversely affect the results of any data analysis and skew conclusions if not handled properly. Statistical analysis is sometimes used to weed the noise out of noisy data. 

*   Data = true signal + noise [7]






*   How about < y > and <y_{new}>?
*   How about S(y) and S(y_{new})?

Not necessarily the mean value with noise will be equal to the mean value without noise. However, if we control the noise so that it follows a uniform distribution ranging from -x to x, then we will be able to keep the average value the same even after adding noise. This is because we will keep the mean of noisy data as 0. Thus, for this work, we chose to insert a noise that follows a uniform distribution between -3 and 3.

Regarding the variance, even if the noise is controlled so that the mean remains the same, there are not a guarantee for the maintenance of the variance.

**Description of code:**

The code below uses **seed = 1** as seed to be used in four generators:

*   generator_additive_noise;
*   generator_multiplicative_noise;
*   generator_multiplicative_additive_noise;
*   generator_uniform.

All generators are of type Permuted Congruential Generator (64-bit, PCG64) [5], and have a period of $2^{128}$.

The "numpy.random.Generator.integers" method [6] returns random integers from a **"discrete uniform" distribution**. The first three generators are used to generate random integers in the range [-3, 3] and the last generator is used to generate random integers in the range [0, len(x_domain)], that is, it is used to generate a position in the array 'x_domain'.

In the code below, functions are defined for inserting additive, multiplicative, additive and multiplicative noise.

```python
seed = 1
generator_additive_noise = Generator(PCG64(seed))
generator_multiplicative_noise = Generator(PCG64(seed))
generator_multiplicative_additive_noise = Generator(PCG64(seed))
generator_uniform = Generator(PCG64(seed))

#Insertion of additive noise
def f_real_additive_noise(x):
  noise = 0
  #If the value drawn is 0, draw again, because adding 0 is not introducing noise
  while noise == 0:
    #Draws a random number between -3 and 3
    noise = generator_additive_noise.integers(low=-3, high=3, endpoint=True)
  #Noise introduction  
  return f_real(x) + noise 

#Insertion of multiplicative noise
def f_real_multiplicative_noise(x):
  noise = 1
  #If the value drawn is 1, draw again, because multiplying 1 is not introducing noise
  while noise == 1:
    #Draws a random number between -3 and 3
    noise = generator_multiplicative_noise.integers(low=-3, high=3, endpoint=True)
  #Noise introduction  
  return f_real(x) * noise 

#Insertion of multiplicative and additive noise
def f_real_multiplicative_additive_noise(x):
  noise_1 = 1
  noise_2 = 0
  #If you don't introduce noise, draw again
  while (noise_1 == 1) and (noise_2 == 0):
    #Draws two random number between -3 and 3
    noise_1 = generator_multiplicative_additive_noise.integers(low=-3, high=3, endpoint=True)
    noise_2 = generator_multiplicative_additive_noise.integers(low=-3, high=3, endpoint=True)
  #Noise introduction  
  return f_real(x) * noise_1 + noise_2     

def introduce_noise(x_domain, delta, f_noise):
  y = []
  for x in x_domain:
      y.append(f_real(x))
  count = 0
  pos_array = []
  while len(pos_array) < int(delta * len(x_domain)):
    pos = generator_uniform.integers(low=0, high=len(x_domain))
    if (pos not in pos_array):
      y[pos] = f_noise(x_domain[pos])
      pos_array.append(pos)  
  return y  
```

**Description of code:**

The code below shows the insertion of the three types of noise (additive, multiplicative, additive and multiplicative) with the following percentages:

*   5%
*   25%
*   50%
*   75%

```python
#For 5% noise
y_additive_noise_5perc = introduce_noise(x_domain, 0.05, f_real_additive_noise)
y_multiplicative_noise_5perc = introduce_noise(x_domain, 0.05, f_real_multiplicative_noise)
y_multiplicative_additive_noise_5perc = introduce_noise(x_domain, 0.05, f_real_multiplicative_additive_noise)

#For 25% noise
y_additive_noise_25perc = introduce_noise(x_domain, 0.25, f_real_additive_noise)
y_multiplicative_noise_25perc = introduce_noise(x_domain, 0.25, f_real_multiplicative_noise)
y_multiplicative_additive_noise_25perc = introduce_noise(x_domain, 0.25, f_real_multiplicative_additive_noise)

#For 50% noise
y_additive_noise_50perc = introduce_noise(x_domain, 0.5, f_real_additive_noise)
y_multiplicative_noise_50perc = introduce_noise(x_domain, 0.5, f_real_multiplicative_noise)
y_multiplicative_additive_noise_50perc = introduce_noise(x_domain, 0.5, f_real_multiplicative_additive_noise)

#For 75% noise
y_additive_noise_75perc = introduce_noise(x_domain, 0.75, f_real_additive_noise)
y_multiplicative_noise_75perc = introduce_noise(x_domain, 0.75, f_real_multiplicative_noise)
y_multiplicative_additive_noise_75perc = introduce_noise(x_domain, 0.75, f_real_multiplicative_additive_noise)
```
