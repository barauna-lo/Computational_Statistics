**Box-Muller for generating normally distributed random numbers**

The Box-Muller transform starts with 2 random uniform numbers u and v - Generate an exponentially distributed variable r2 from u using the inverse transform method - This means that r is an exponentially distributed variable on (0,) - Generate a variable  uniformly distributed on (0,2) from v by scaling - In polar coordinates, the vector (r,) has an independent bivariate normal distribution - Hence the projection onto the x and y axes give independent univariate normal random numbers

```python
n = 1000
u1 = np.random.random(n)
u2 = np.random.random(n)
r_squared = -2*np.log(u1)
r = np.sqrt(r_squared)
theta = 2*np.pi*u2
x = r*np.cos(theta)
y = r*np.sin(theta)
```

# Exercise week 2: Linear regression

Linear relation between two variables 

Y=0+1X1+

Simple linear regression consists of generating a regression model (equation of a straight line) that allows explaining the linear relationship that exists between two variables. The dependent or response variable is identified as Y and the predictor or independent variable as X.


In some cases, the population 0 and 1 values are unknown, so their estimates ^0 and ^1 are obtained from a sample. These estimates are known as regression coefficients or **least square coefficient estimates**  , since they take those values that minimize the residual sum of squares, giving rise to the line that passes closest to all the points. 


image.png

image.png

```python
def estimate_coef(x, y):
    # number of observations/points
    n = np.size(x)
 
    # mean of x and y vector
    m_x = np.mean(x)
    m_y = np.mean(y)
 
    # calculating standar deviation about x

    SS_xy = np.sum(y*x) - n*m_y*m_x
    SS_xx = np.sum(x*x) - n*m_x*m_x
 
    # calculating regression coefficients
    
    b_1 = SS_xy / SS_xx
    b_0 = m_y - b_1*m_x
 
    return (b_0, b_1)
 
def plot_regression_line(x, y, b):
    # plotting the actual points as scatter plot
    plt.scatter(x, y, color = "m",
               marker = "o", s = 30)
 
    # predicted response vector
    y_pred = (b[0] + b[1]*x)
 
    # plotting the regression line
    plt.plot(x, y_pred, color = "g")
 
    # putting labels
    plt.xlabel('x')
    plt.ylabel('y')
 
    # function to show plot
    plt.show()
 
def main():
    # observations / data
    x = np.random.randn(100)
    y = np.random.randn(100)
 
    # estimating coefficients
    b = estimate_coef(x, y)
    print("Estimated coefficients:\nb_0 = {}  \
          \nb_1 = {}".format(b[0], b[1]))
 
    # plotting regression line
    plot_regression_line(x, y, b)
 
if __name__ == "__main__":
    main()
```

```python

# Train set
a=2
b=3
trX = np.linspace(-2, 2, 101)
trY = b + a * trX + np.random.randn(*trX.shape) * 0.33

# initial parameters
num_steps = 100
learningRate = 0.10
criteria = 1e-8
b_0 = 1
b_1 = 1

# Iterative gradient

for step in range(0, num_steps):
    b_0_gradient = 0
    b_1_gradient = 0
    N = float(len(trX))
    for i in range(0, len(trX)):
        b_0_gradient -= (2/N) * (trY[i] - (b_0 + b_1 * trX[i]))
        b_1_gradient -= (2/N) * (trY[i] - (b_0 + b_1 * trX[i])) * trX[i]
        
    b_0 = b_0 - (learningRate * b_0_gradient)
    b_1 = b_1 - (learningRate * b_1_gradient)
    if max(abs(learningRate * b_0_gradient), abs(learningRate * b_1_gradient)) < criteria:
        break
    
# Results
print("Values:", b_0, b_1, "steps", step)
```

```python
from scipy.stats import t, norm
# Paso 1
m = 1000                         #Tamao de la muestra observada.
mo = np.random.randn(m)          #Generacin de una muestra N(0,1) de tamao m.
est_pi = np.mean                 #Estimador plug-in.
alpha = 0.05                     #Nivel de significancia.

def intervalo_Bootstrap(x, estimador, N, alpha = 0.05):
    # Paso 2 y Paso 3
    thetas_est = [estimador(np.random.choice(x, m, replace = True)) for j in range(N)]
    
    # Paso 4
    muestra_ord = sorted(thetas_est)
    
    # Paso 5
    l = np.ceil(alpha/2*N)
    u = np.ceil((1-alpha/2)*N)

    # Paso 6
    t = estimador(x)
    
    #Paso 7.
    interval = [2*t - muestra_ord[int(u)], 2*t - muestra_ord[int(l)]]
    return interval


#Repeticiones
N = 1000

# Muestra aleatoria
x = [np.random.randn(m) for k in range(100)]

# Intervalos bootstrap
boot_inf = [intervalo_Bootstrap(x[k], np.mean, N, alpha)[0] for k in range(100)]
boot_sup = [intervalo_Bootstrap(x[k], np.mean, N, alpha)[1] for k in range(100)]

# Intervalos tericos
ex_inf = [np.mean(x[k])-t.ppf(1-alpha/2, m-1)*np.std(x[k])/np.sqrt(len(x[k])) for k in range(100)]
ex_sup = [np.mean(x[k])+t.ppf(1-alpha/2, m-1)*np.std(x[k])/np.sqrt(len(x[k])) for k in range(100)]

# Graficamos ambos intervalos para ver como "convergen"
plt.figure(figsize = (16,8))
plt.title('Intervalo Bootstrap BCa')
plt.hlines(y = 0, xmin = 0, xmax = 100, linestyles = 'solid', colors = 'indigo', label = 'Media terica.')
plt.plot(ex_sup, 's', ls = 'dashed', c = 'mediumturquoise', label = 'Intervalo Terico')
plt.plot(ex_inf, 's', ls = 'dashed', c = 'mediumturquoise')
plt.plot(boot_inf, 's', ls = 'dashed', c = 'lightcoral', label = 'Intervalo Bootstrap')
plt.plot(boot_sup, 's', ls = 'dashed', c = 'lightcoral')
plt.legend(loc = 0)
plt.show()
```

```python
#  number of partitions of a set with n elements into k subsets
def countP(n, k):
    # Base cases
    if (n == 0 or k == 0 or k > n):
        return 0
    if (k == 1 or k == n):
        return 1
     
    # S(n+1, k) = k*S(n, k) + S(n, k-1)
    return (k * countP(n - 1, k) +
                countP(n - 1, k - 1))
 
   # print(countP(25, 3))
```

```python
# Permutation Coefficient P(n, k)
def permutationCoeff(n, k):
	fact = [0 for i in range(n + 1)]

	# base case
	fact[0] = 1

	# Calculate value
	# factorials up to n
	for i in range(1, n + 1):
		fact[i] = i * fact[i - 1]

	# P(n, k) = n!/(n-k)!
	return int(fact[n] / fact[n - k])
```

```python
#number of k-element subsets (or k-combinations) of a n-element set.
# A naive recursive Python implementation


def binomialCoeff(n, k):

	if k > n:
		return 0
	if k == 0 or k == n:
		return 1

	# Recursive Call
	return binomialCoeff(n-1, k-1) + binomialCoeff(n-1, k)
```

```python
def partition(collection):
    if len(collection) == 1:
        yield [ collection ]
        return

    first = collection[0]
    for smaller in partition(collection[1:]):
        # insert `first` in each of the subpartition's subsets
        for n, subset in enumerate(smaller):
            yield smaller[:n] + [[ first ] + subset]  + smaller[n+1:]
        # put `first` in its own subset 
        yield [ [ first ] ] + smaller
```
